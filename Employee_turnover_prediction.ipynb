```python
df = df.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis = 1)
```

```python
sns.histplot(data=df, x='BusinessTravel', element='step', color='purple', stat='percent')
```

```python
attrition_dummies = pd.get_dummies(df['Attrition'])
attrition_dummies.head()
```

```python
df = pd.concat([df,attrition_dummies],axis=1)
df.head()
```

```python
df = df.drop(['Attrition','No'],axis = 1)
df.head()
```

```python
df = df.drop(['Age','JobLevel'],axis = 1)
```

Data Processing


```python
from sklearn.preprocessing import LabelEncoder
for column in df.columns:
    if df[column].dtype==np.number:
        continue
    else:
      df[column]=LabelEncoder().fit_transform(df[column])
```

Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10,criterion = 'entropy',random_state=0)
```

```python
x= df.drop(['Yes'],axis = 1)
y = df['Yes']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 0)
```

```python
x_train.head()
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)
```

```python
pred = rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred)
```
